```python
import matplotlib.pyplot as plt
import sys, os
import Players

sys.stdout = open(os.devnull, 'w')
```

```python
num = 500
```

```python
winners = {}
players = ["1","4"]
for i in range(0,num):
    import monopoly
    winner = monopoly.playGame(players)
    if winner in winners:
        winners[winner] += 1
    else:
        winners[winner] = 1
    del sys.modules['monopoly']

lists = sorted(winners.items())
x, y = zip(*lists)

fig, ax = plt.subplots()
ax.pie(y, labels=x, autopct='%.1f%%')
plt.show()
```
